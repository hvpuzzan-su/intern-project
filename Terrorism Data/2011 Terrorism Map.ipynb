```python
import folium
from folium import plugins
import pandas as pd
import branca
from folium.plugins import HeatMap
import numpy as np
import re

data = pd.read_csv('jt_dataset.txt' , index_col=['Event ID'])
df = pd.DataFrame(data)
df['Event Date'] = pd.to_datetime(df['Event Date'])
df['Year'] = df['Event Date'].dt.year

def significance_to_numeric(x):
    if x == 'Common':
        return 0
    if x == 'Notable':
        return 1
    if x == 'Significant':
        return 2
    if x == 'Critical':
        return 3
df['Significance Score'] = df['Significance'].apply(significance_to_numeric)

df2011 = df[df.Year == 2011]

map1 = folium.Map(location=[df2011['Latitude'].mean(), df2011['Longitude'].mean()], zoom_start=2)

for i in range (0, len(df2011)):
    significance = df2011['Significance Score'].iloc[i]
    if significance == 0:
        folium.CircleMarker([df2011['Latitude'].iloc[i], df2011['Longitude'].iloc[i]], 
                            radius = 3, 
                            fill = True, 
                            fill_color = 'yellow', 
                            popup = ("Country: {Country}<br>" "Significance: {Significance}<br>" "Event Description: {Title}"
                                    ).format(Country = re.sub(r"[^a-zA-Z0-9_]+", ' ',df2011['Country'].iloc[i]), 
                                                 Significance = df2011['Significance'].iloc[i], 
                                                 Title = re.sub(r"[^a-zA-Z0-9_]+", ' ', df2011['Title'].iloc[i])),
                            color = 'yellow', 
                            opacity = 0.6, 
                            fill_opacity = 0.6).add_to(map1)
    elif significance == 1:
        folium.CircleMarker([df2011['Latitude'].iloc[i], df2011['Longitude'].iloc[i]], 
                            radius = 3, 
                            fill = True, 
                            fill_color = 'orange', 
                            popup = ("Country: {Country}<br>" "Significance: {Significance}<br>" "Event Description: {Title}"
                                    ).format(Country = re.sub(r"[^a-zA-Z0-9_]+", ' ',df2011['Country'].iloc[i]), 
                                                 Significance = df2011['Significance'].iloc[i], 
                                                 Title = re.sub(r"[^a-zA-Z0-9_]+", ' ', df2011['Title'].iloc[i])),
                            color = 'orange', 
                            opacity = 0.6, 
                            fill_opacity = 0.6).add_to(map1)
    elif significance == 2:
        folium.CircleMarker([df2011['Latitude'].iloc[i], df2011['Longitude'].iloc[i]], 
                            radius = 3, 
                            fill = True, 
                            fill_color = 'red', 
                            popup = ("Country: {Country}<br>" "Significance: {Significance}<br>" "Event Description: {Title}"
                                    ).format(Country = re.sub(r"[^a-zA-Z0-9_]+", ' ',df2011['Country'].iloc[i]), 
                                                 Significance = df2011['Significance'].iloc[i], 
                                                 Title = re.sub(r"[^a-zA-Z0-9_]+", ' ', df2011['Title'].iloc[i])),
                            color = 'red', 
                            opacity = 0.6, 
                            fill_opacity = 0.6).add_to(map1)
    else:
        folium.CircleMarker([df2011['Latitude'].iloc[i], df2011['Longitude'].iloc[i]], 
                            radius = 3, 
                            fill = True, 
                            fill_color = 'darkred', 
                            popup = ("Country: {Country}<br>" "Significance: {Significance}<br>" "Event Description: {Title}"
                                    ).format(Country = re.sub(r"[^a-zA-Z0-9_]+", ' ',df2011['Country'].iloc[i]), 
                                                 Significance = df2011['Significance'].iloc[i], 
                                                 Title = re.sub(r"[^a-zA-Z0-9_]+", ' ', df2011['Title'].iloc[i])),
                            color = 'darkred', 
                            opacity = 0.6, 
                            fill_opacity = 0.6).add_to(map1)
map1.save("C:\\Users\\hpuzzang\\Documents\\Terrorism Maps\\2011_popup_map.html")
```
